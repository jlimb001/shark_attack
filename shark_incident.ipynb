```python
# Renames all instances of 'Surfing' or 'surfing' into 'surfing'
def rename_surfing(df, column_name):
    # First, check if the column exists in the DataFrame
    if column_name not in df.columns:
        print(f"Column '{column_name}' does not exist in the DataFrame.")
        return df

    # Replace all variations of 'surfing' with 'surfing' and remove any other text
    df[column_name] = df[column_name].str.replace(r'\b.*surfing.*\b', 'surfing', case=False, regex=True)

    # Remove any extra spaces
    df[column_name] = df[column_name].str.strip()

    return df

# Example usage
df = rename_surfing(df, 'Activity')

# Get the unique values and their counts
activity_counts = df['Activity'].value_counts()

# Print the activity 'surfing' and how many times it occurs
if 'surfing' in activity_counts:
    print(f"surfing: {activity_counts['surfing']} times")
else:
    print("No activities with 'surfing' found.")
```

```python
# I created a function to rename all the cells of the column activity with surf or surfrelated subjects turn into surfing

import pandas as pd

def rename_surfing_related_terms(df, column_name):
    # List of terms related to 'surfing'
    surfing_related_terms = [
        'Surf-sking', 'Surf-skiing', 'Surfng', 'surfski', 'Surf paddling',
        'Scurfing', 'Surf-fishing', 'Fishing from surfski', 'Surfboard',
        'Lying on surfboard', 'Paddling on surfboard or body board',
        'Lying on surfboard & paddling', 'Resting on surfboard',
        'Jumped off surfboard & landed on the shark', 'Fell off surfboard 20\' from shore',
        'Lying atop surfboard', 'Surf paddling', 'Scurfing (surfboard being towed behind a boat)',
        'Fishing from Surfboard', 'Swimming, towing surfboard',
        'Wading with surfboard', 'Standing alongside surfboard',
        'Playing in the surf with his 2 dogs', 'Chasing shark out of bathing area while riding on a surf-ski',
        'Pushed surfmat of a young girl out of the shark\'s path, drawing shark\'s attention to his own board'
    ]


    df[column_name] = df[column_name].replace(surfing_related_terms, 'surfing')

    return df

df = rename_surfing_related_terms(df, 'Activity')


surfing_count = (df['Activity'] == 'surfing').sum()
print(f"'surfing' appears {surfing_count} times in the 'Activity' column.")

```

```python
# When an entry is missing and the corresponding "Injury" column contains keywords like "bit", "bitten", "shark", or "bite", the species is assumed to be a shark.#

def override_species_with_shark(df_surfing_only):
    # Apply the condition and set 'Species' to 'shark' where it is NaN and 'Injury' contains certain keywords
    df_surfing_only.loc[
        df_surfing_only['Species '].isna() &
        df_surfing_only['Injury'].str.contains(r'\b(bit|bitten|shark|bite)\b', case=False, na=False),
        'Species '
    ] = 'shark'
    return df_surfing_only
```

```python
df_surfing_only['Time'].replace({'?': pd.NaT, 'NaN': pd.NaT}, inplace=True)

# Convert 'Time' column to datetime, coercing errors into NaT
df_surfing_only['Time'] = pd.to_datetime(df_surfing_only['Time'], errors='coerce', format='%H:%M')

# Replace NaT values with '00:00' time while preserving the original date
df_surfing_only['Time'] = df_surfing_only['Time'].where(df_surfing_only['Time'].notna(),
                                                         df_surfing_only['Time'].apply(lambda x: x.replace(hour=0, minute=0, second=0, microsecond=0)))

# Convert datetime to just time (remove date part)
df_surfing_only['Time'] = df_surfing_only['Time'].dt.time

# Format the time for display (HH:MM)
df_surfing_only['Time_display'] = df_surfing_only['Time'].apply(lambda x: x.strftime('%H:%M') if isinstance(x, pd.Timestamp) else '00:00')

# Verify the result
print("Time column with time dtype (without date):")
print(df_surfing_only['Time'])

print("\nTime for display (HH:MM):")
print(df_surfing_only['Time_display'])
```
